# 資料清理與格式化(json)

```python
def cleanJsonData(file_list_before,file_list_after):
    #先開啟要寫的新檔案
    with open(file_list_after,'w') as cleanData:
        #再開啟舊的檔案
        with open(file_list_before,'r') as rowdata:
            # 檢查用的count不重要
            count = 0
            count_space = 0
            # 將乾淨的data都放進這個list中
            clean_rowData = []
            # 逐行讀取rowdata
            for line in rowdata:
                # 檢查每一行是否為空行，如果有空行就印出提示字元並跳過不儲存在list
                if line =="\n":
                    count_space+=1
                    print("這裡有空行")
                else:
                    # 上面檢查過是否空行，只要不是空行就會把資料存入list
                    clean_rowData.append(line)
                count+=1
            # 臨時的count變數，用來數乾淨資料的數量並把中括號加在頭尾
            tmpCount = 1
            # 印出乾淨的資料大概有多少行
            print(len(clean_rowData))
            for i in range(len(clean_rowData)):
                # 如果count大於等於乾淨資料的數量就加上後中括號
                if tmpCount >= len(clean_rowData):
                    clean_rowData.append(']')
                # 如果該筆資料還不是最後一筆那就將該筆資料的換行取代成逗號加換行
                else:
                    clean_rowData[i]=clean_rowData[i].replace("\n",",\n")
                tmpCount+=1
        # 最後將資料的頭加上前中括號和換行
        clean_rowData.insert(0,'[\n')
        # 將資料寫入新檔案
        for lines in clean_rowData:
            cleanData.writelines(lines)
```

```python
# 匯入os套件
import os
# 將rowdata位置存入path變數
path='D:\\桌面\\rowData\\'
# 將處理後的資料位置存入path_new變數
path_new='D:\\桌面\\after_Clean_data\\'
# 讀取path位置內的所有檔案並存成list
file_list = os.listdir(path)
# for迴圈抓檔案list中的每一個元素

for i in file_list:
    # 將檔名用split切割出只有日期的部分，當作之後新的檔名要用的變數
    split_file_name = i.split('.')
    # 將上面處理好的日期結合新的路徑輸出
    split_file_name_after = path_new+split_file_name[0]+'_after.json'
    # 將舊檔名與舊的路徑結合
    split_file_name_before = path+i
    # 呼叫上面定義的function處理空行問題並且將每一欄資料都加上逗點
    cleanJsonData(split_file_name_before,split_file_name_after)
    #印出舊的檔名，提示程式進度
    print(split_file_name_before)
    
    
```
